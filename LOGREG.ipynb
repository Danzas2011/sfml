# Функция предобработки

```python
def preproc_data(df_input):
    df_output = df_input.copy()
    
    #df_output=df_output.drop(['age'], axis=1)
    
    #df_output['age_0_19'] = df_output['age'] < 20
    #df_output['age_20_29'] = (df_output['age'] >= 20) & (df['age'] < 30)
    #df_output['age_30_39'] = (df_output['age'] >= 30) & (df['age'] < 40)
    #df_output['age_40_49'] = (df_output['age'] >= 40) & (df['age'] < 50)
    #df_output['age_50_59'] = (df_output['age'] >= 50) & (df['age'] < 60)
    #df_output['age_60_69'] = (df_output['age'] >= 60) & (df['age'] < 70)
    #df_output['age_70_79'] = (df_output['age'] >= 70) & (df['age'] < 80)
    #df_output['age_80_99'] = (df_output['age'] >= 80) & (df['age'] < 90)

    #df_output['age_0_19'] = df_output['age_0_19'].astype(int, inplace = True)
    #df_output['age_20_29'] = df_output['age_20_29'].astype(int, inplace = True)
    #df_output['age_30_39'] = df_output['age_30_39'].astype(int, inplace = True)
    #df_output['age_40_49'] = df_output['age_40_49'].astype(int, inplace = True)
    #df_output['age_50_59'] = df_output['age_50_59'].astype(int, inplace = True)
    #df_output['age_60_69'] = df_output['age_60_69'].astype(int, inplace = True)
    #df_output['age_70_79'] = df_output['age_70_79'].astype(int, inplace = True)
    #df_output['age_80_99'] = df_output['age_80_99'].astype(int, inplace = True)
    
    #df_output=df_output.drop(['age'], axis=1)
    
    #df_output['f1'] = df_output['previous'] + df_output['campaign']
    #df_output['f1'] = (df_output['f1'] - df_output['f1'].min())/(df_output['f1'].max()-df_output['f1'].min())
    #df_output=df_output.drop(['previous','campaign'], axis=1)
    
    #df_output['campaign_bool'] = df_output['campaign'] > 15
    df_output['campaign_bool'] = df_output['campaign'] < df_output['campaign'].mean()
    df_output['campaign_bool'] = df_output['campaign_bool'].astype(int, inplace = True)
    #df_output=df_output.drop(['campaign'], axis=1)
    
    
    # добавить масштабирование и дискретизацию признаков
    #df_output['age_scale'] = (df_output['age'] - df_output['age'].min())/(df_output['age'].max()-df_output['age'].min())
    df_output['duration_scale'] = (df_output['duration'] - df_output['duration'].min())/(df_output['duration'].max()-df_output['duration'].min())
    #df_output['campaign_scale'] = (df_output['campaign'] - df_output['campaign'].min())/(df_output['campaign'].max()-df_output['campaign'].min())
    #df_output['pdays_scale'] = (df_output['pdays'] - df_output['pdays'].min())/(df_output['pdays'].max()-df_output['pdays'].min())
    #df_output['previous_scale'] = (df_output['previous'] - df_output['previous'].min())/(df_output['previous'].max()-df_output['previous'].min())
    
    #df_output['emp.var.rate_scale'] = (df_output['emp.var.rate'] - df_output['emp.var.rate'].min())/(df_output['emp.var.rate'].max()-df_output['emp.var.rate'].min())
    #df_output['cons.price.idx_scale'] = (df_output['cons.price.idx'] - df_output['cons.price.idx'].min())/(df_output['cons.price.idx'].max()-df_output['cons.price.idx'].min())
    #df_output['cons.conf.idx_scale'] = (df_output['cons.conf.idx'] - df_output['cons.conf.idx'].min())/(df_output['cons.conf.idx'].max()-df_output['cons.conf.idx'].min())
    #df_output['euribor3m_scale'] = (df_output['euribor3m'] - df_output['euribor3m'].min())/(df_output['euribor3m'].max()-df_output['euribor3m'].min())
           
    df_output['duration_bool'] = df_output['duration'] < df_output['duration'].mean()
    df_output['duration_bool'] = df_output['duration_bool'].astype(int, inplace = True)
    #df_output=df_output.drop(['duration'], axis=1)
    
    
    #df_output=df_output.drop(['age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate',
                              #'cons.price.idx', 'cons.conf.idx', 'euribor3m' ], axis=1)
    
    #df_output=df_output.drop([ 'duration'], axis=1)
    
    #df_output['duration'] = df_output['duration']**3
    
    #df_output['emp.var.rate'] = df_output['emp.var.rate']**2
    df_output['emp.var.rate.mean'] = df_output['emp.var.rate'] < df_output['emp.var.rate'].mean()
    df_output['emp.var.rate.mean'] = df_output['emp.var.rate.mean'].astype(int, inplace = True)
    #df_output=df_output.drop(['emp.var.rate'], axis=1)
    
    #df_output['cons.price.idx'] = df_output['cons.price.idx']**2
    #df_output['cons.conf.idx'] = df_output['cons.conf.idx']**2
    #df_output['euribor3m'] = df_output['euribor3m']**2
    #df_output['euribor3m_bool'] = df_output['euribor3m'] > 1
    df_output['euribor3m_bool'] = df_output['euribor3m'] < df_output['euribor3m'].mean()
    df_output['euribor3m_bool'] = df_output['euribor3m_bool'].astype(int, inplace = True)
    #df_output=df_output.drop(['euribor3m'], axis=1)
    
    df_output['nr.employed_mean'] = df_output['nr.employed'] < df_output['nr.employed'].mean()
    df_output['nr.employed_mean'] = df_output['nr.employed_mean'].astype(int, inplace = True)
    #df_output=df_output.drop(['nr.employed'], axis=1)
    
    
    #df_output=df_output.drop(['emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m'], axis=1)
    df_output=df_output.drop(['cons.price.idx','cons.conf.idx'], axis=1)
    
    #for i in ['PERSONAL_INCOME', 'CREDIT', 'FST_PAYMENT', 'LOAN_AVG_DLQ_AMT', 'LOAN_MAX_DLQ_AMT']:
        #if str(df[i].dtype) == 'object':
            #df_output[i]=df_output[i].map(lambda x: x.replace(',', '.')).astype('float')
      
    
    #df_output = df_output.fillna(df_output.mean())
    #df_output.CREDIT[df_output.CREDIT.isnull()] = df_output.CREDIT.mean()
    #df_output = df_output.fillna(0)
         
    text_features = [col for col in df_output.columns if (df_output[col].dtype == 'object' and str(col)!='sample') and str(col)!='_id']    
    
    #data_describe = df_output.describe(include=[object])
    #for c in text_features:
        #df_output[c] = df_output[c].fillna(data_describe[c]['top'])
        #df_output[c] = df_output[c].fillna('ПУСТО') 
    
    text_data = pd.get_dummies(df_output[text_features], prefix = text_features, drop_first = False)
    
    numeric_features = [col for col in df_output.columns if df_output[col].dtype != 'object']
   
    df_output = pd.concat([df_output['_id'], df_output[numeric_features], text_data, df_output['sample'] ], axis = 1)
    
    return df_output
```

```python
from sklearn.metrics import roc_auc_score
def gini_score(y_true, y_pred):
    return -1 + 2 * roc_auc_score(y_true, y_pred)
```
